```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
COLORS = ['#E63946', '#457B9D', '#2A9D8F', '#E9C46A', '#F4A261']

df = pd.read_csv('Mall_Customers.csv')
print("Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print(df.head())
```

```python
# EDA
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Exploratory Data Analysis', fontsize=16, fontweight='bold')

axes[0, 0].hist(df['Age'], bins=20, color='#457B9D', edgecolor='white')
axes[0, 0].set_title('Age Distribution')
axes[0, 0].set_xlabel('Age')

axes[0, 1].hist(df['Annual Income (k$)'], bins=20, color='#2A9D8F', edgecolor='white')
axes[0, 1].set_title('Annual Income Distribution')
axes[0, 1].set_xlabel('Annual Income (k$)')

axes[0, 2].hist(df['Spending Score (1-100)'], bins=20, color='#E63946', edgecolor='white')
axes[0, 2].set_title('Spending Score Distribution')
axes[0, 2].set_xlabel('Spending Score (1-100)')

gender_counts = df['Gender'].value_counts()
axes[1, 0].bar(gender_counts.index, gender_counts.values, color=['#F4A261', '#457B9D'])
axes[1, 0].set_title('Gender Distribution')

axes[1, 1].scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'],
                   alpha=0.6, color='#457B9D', edgecolors='white', linewidth=0.5)
axes[1, 1].set_title('Income vs Spending Score')
axes[1, 1].set_xlabel('Annual Income (k$)')
axes[1, 1].set_ylabel('Spending Score')

axes[1, 2].scatter(df['Age'], df['Spending Score (1-100)'],
                   alpha=0.6, color='#2A9D8F', edgecolors='white', linewidth=0.5)
axes[1, 2].set_title('Age vs Spending Score')
axes[1, 2].set_xlabel('Age')
axes[1, 2].set_ylabel('Spending Score')

plt.tight_layout()
plt.savefig('eda_plots.png', dpi=150, bbox_inches='tight')
plt.show()
print("EDA done! eda_plots.png saved.")
```

```python
# Preprocessing + Elbow Method
le = LabelEncoder()
df['Gender_encoded'] = le.fit_transform(df['Gender'])

X = df[['Annual Income (k$)', 'Spending Score (1-100)']].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertias = []
silhouettes = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)
    silhouettes.append(silhouette_score(X_scaled, km.labels_))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Optimal K Selection', fontsize=14, fontweight='bold')

axes[0].plot(K_range, inertias, 'o-', color='#457B9D', linewidth=2, markersize=8)
axes[0].axvline(x=5, color='#E63946', linestyle='--', label='Optimal K=5')
axes[0].set_title('Elbow Method — Inertia vs K')
axes[0].set_xlabel('Number of Clusters (K)')
axes[0].set_ylabel('Inertia')
axes[0].legend()

axes[1].plot(K_range, silhouettes, 's-', color='#2A9D8F', linewidth=2, markersize=8)
axes[1].axvline(x=5, color='#E63946', linestyle='--', label='Optimal K=5')
axes[1].set_title('Silhouette Score vs K')
axes[1].set_xlabel('Number of Clusters (K)')
axes[1].set_ylabel('Silhouette Score')
axes[1].legend()

plt.tight_layout()
plt.savefig('elbow_silhouette.png', dpi=150, bbox_inches='tight')
plt.show()
print(f"Optimal K = {K_range[np.argmax(silhouettes)]}")
print(f"Best silhouette score = {max(silhouettes):.4f}")
```

```python
# K-Means Clustering
OPTIMAL_K = 5
kmeans = KMeans(n_clusters=OPTIMAL_K, random_state=42, n_init=10)
df['KMeans_Cluster'] = kmeans.fit_predict(X_scaled)

kmeans_sil = silhouette_score(X_scaled, df['KMeans_Cluster'])
print(f"K-Means Silhouette Score: {kmeans_sil:.4f}")

fig, ax = plt.subplots(figsize=(10, 7))
for cluster_id in range(OPTIMAL_K):
    mask = df['KMeans_Cluster'] == cluster_id
    ax.scatter(df.loc[mask, 'Annual Income (k$)'],
               df.loc[mask, 'Spending Score (1-100)'],
               c=COLORS[cluster_id], label=f'Cluster {cluster_id}',
               s=80, alpha=0.8, edgecolors='white', linewidth=0.5)

centers_orig = scaler.inverse_transform(kmeans.cluster_centers_)
ax.scatter(centers_orig[:, 0], centers_orig[:, 1],
           c='black', marker='X', s=200, zorder=5, label='Centroids')

ax.set_title('K-Means Customer Segmentation (K=5)', fontsize=14, fontweight='bold')
ax.set_xlabel('Annual Income (k$)')
ax.set_ylabel('Spending Score (1-100)')
ax.legend()
plt.tight_layout()
plt.savefig('kmeans_clusters.png', dpi=150, bbox_inches='tight')
plt.show()
print("K-Means done! kmeans_clusters.png saved.")
```

```python
# Hierarchical Clustering
fig, ax = plt.subplots(figsize=(14, 6))
Z = linkage(X_scaled, method='ward')
dendrogram(Z, ax=ax, truncate_mode='lastp', p=20, leaf_rotation=45,
           color_threshold=6, above_threshold_color='#888')
ax.set_title('Hierarchical Clustering Dendrogram (Ward Linkage)', fontsize=14, fontweight='bold')
ax.set_xlabel('Samples')
ax.set_ylabel('Distance')
ax.axhline(y=6, color='#E63946', linestyle='--', label='Cut → 5 clusters')
ax.legend()
plt.tight_layout()
plt.savefig('dendrogram.png', dpi=150, bbox_inches='tight')
plt.show()

agglo = AgglomerativeClustering(n_clusters=5, linkage='ward')
df['Agglo_Cluster'] = agglo.fit_predict(X_scaled)
agglo_sil = silhouette_score(X_scaled, df['Agglo_Cluster'])
print(f"Hierarchical Silhouette Score: {agglo_sil:.4f}")

fig, ax = plt.subplots(figsize=(10, 7))
for cluster_id in range(5):
    mask = df['Agglo_Cluster'] == cluster_id
    ax.scatter(df.loc[mask, 'Annual Income (k$)'],
               df.loc[mask, 'Spending Score (1-100)'],
               c=COLORS[cluster_id], label=f'Cluster {cluster_id}',
               s=80, alpha=0.8, edgecolors='white', linewidth=0.5)
ax.set_title('Hierarchical Clustering — Customer Segments', fontsize=14, fontweight='bold')
ax.set_xlabel('Annual Income (k$)')
ax.set_ylabel('Spending Score (1-100)')
ax.legend()
plt.tight_layout()
plt.savefig('hierarchical_clusters.png', dpi=150, bbox_inches='tight')
plt.show()
print("Hierarchical done! hierarchical_clusters.png saved.")
```

```python
# Cluster Profiling & Model Comparison
cluster_names = {
    0: 'Careful Spenders',
    1: 'Impulsive Spenders',
    2: 'Standard Customers',
    3: 'Affluent Savers',
    4: 'Target Customers',
}

profile = df.groupby('KMeans_Cluster').agg(
    Count=('CustomerID', 'count'),
    Avg_Age=('Age', 'mean'),
    Avg_Income=('Annual Income (k$)', 'mean'),
    Avg_Spending=('Spending Score (1-100)', 'mean'),
    Female_Pct=('Gender', lambda x: (x == 'Female').mean() * 100)
).round(1)

profile.index = [cluster_names.get(i, f'Cluster {i}') for i in profile.index]
print("=" * 55)
print("CLUSTER PROFILES")
print("=" * 55)
print(profile.to_string())

fig, ax = plt.subplots(figsize=(10, 5))
profile_norm = profile.drop('Count', axis=1).apply(
    lambda x: (x - x.min()) / (x.max() - x.min())
)
sns.heatmap(profile_norm, annot=profile.drop('Count', axis=1),
            fmt='.1f', cmap='YlOrRd', ax=ax,
            linewidths=0.5, linecolor='white')
ax.set_title('Customer Segment Profiles', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('cluster_profiles_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n" + "=" * 55)
print("MODEL COMPARISON")
print("=" * 55)
print(f"K-Means Silhouette Score    : {kmeans_sil:.4f}")
print(f"Hierarchical Silhouette Score: {agglo_sil:.4f}")
winner = 'K-Means' if kmeans_sil >= agglo_sil else 'Hierarchical'
print(f"Best performer: {winner}")
```

```python
# Check actual cluster characteristics to assign correct names
actual_profile = df.groupby('KMeans_Cluster').agg(
    Count=('CustomerID', 'count'),
    Avg_Income=('Annual Income (k$)', 'mean'),
    Avg_Spending=('Spending Score (1-100)', 'mean'),
).round(1)
print(actual_profile)
```

```python
# Corrected cluster names + save final output
cluster_names_fixed = {
    0: 'Standard Customers',
    1: 'Target Customers',
    2: 'Impulsive Spenders',
    3: 'Affluent Savers',
    4: 'Careful Spenders',
}

df['Segment'] = df['KMeans_Cluster'].map(cluster_names_fixed)

final_profile = df.groupby('Segment').agg(
    Count=('CustomerID', 'count'),
    Avg_Age=('Age', 'mean'),
    Avg_Income=('Annual Income (k$)', 'mean'),
    Avg_Spending=('Spending Score (1-100)', 'mean'),
    Female_Pct=('Gender', lambda x: (x == 'Female').mean() * 100)
).round(1)

print("=" * 55)
print("FINAL SEGMENT PROFILES")
print("=" * 55)
print(final_profile.to_string())

df.to_csv('customer_segments_output.csv', index=False)

print("\n" + "=" * 55)
print("PROJECT COMPLETE!")
print("=" * 55)
print("Files saved in C:\\dev:")
print("  ✓ eda_plots.png")
print("  ✓ elbow_silhouette.png")
print("  ✓ kmeans_clusters.png")
print("  ✓ dendrogram.png")
print("  ✓ hierarchical_clusters.png")
print("  ✓ cluster_profiles_heatmap.png")
print("  ✓ customer_segments_output.csv")
```
